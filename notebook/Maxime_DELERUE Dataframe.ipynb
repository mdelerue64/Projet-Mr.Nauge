```python
import pandas as pd

# préciser le chemin du fichier à lire
cheminDuFichierCSV = "C:/Users/user/Documents/CNAM/3eA/Nauge/cnam-athon-pronunciation-main/data/dictionary/1757_Buchanan-J.csv" 

# demander à pandas de lire le fichier et garder ce tableau accessible dans une variable
dfDicoBuch = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")

# afficher à l'écran un échantillon du tableau
dfDicoBuch.sample(n=5)
```

```python
def motVedette(mot):
    cheminDuFichierCSV = "./../data/dictionary/Buchanan_PronChar_counts.csv"
    pronchar = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
    TruePron = []
    transfo = ''
    for lettre in mot:
        if (pronchar[pronchar['char']== lettre]['graph'].isnull().values[0]):
            transfo = ''
        else: transfo = pronchar[pronchar['char']== lettre]['graph'].values[0]
        TruePron.append(transfo)
    TruePron = ''.join(TruePron)
    return TruePron
```

```python
motVedette("hĕxaēʹdron")
```

```python
dfDicoBuch["MotVedette"] = None
for i in range(0, len(dfDicoBuch["pron"])):
    dfDicoBuch["MotVedette"][i]=motVedette(dfDicoBuch["pron"][i])
dfDicoBuch
```

```python
def IPA(mot):
    Motnonpresent = []
    cheminDuFichierCSV = "./../data/dataframe/BigiRichSample.csv"
    pronchar = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
    for i in range(0, len(pronchar["ipa"])):
        if (pronchar["hwd"][i] == mot):
            return pronchar["ipa"][i] 
        else : Motnonpresent.append(mot)
        return Motnonpresent
```

```python
IPA("zotica")
```

```python
dfDicoBuch["IPA"] = None
for i in range(0, len(dfDicoBuch["pron"])):
    dfDicoBuch["IPA"][i]=IPA(dfDicoBuch["MotVedette"][i])
    
dfDicoBuch

#Le code ci-dessus marche et me permettrait de récuperer la liste mot qui ne sont pas présent en plus de créer la colonne IPA.
# Malheuresement, il mettait trop de temps à s'exécuter, j'ai donc utiliser un "merge"
```

```python
cheminDuFichierCSV = "./../data/dataframe/BigiRichSample.csv"
bigi = pd.read_csv(cheminDuFichierCSV, delimiter=";", encoding="utf8")
dfDicoBuch_new = dfDicoBuch.merge(bigi, left_on='MotVedette', right_on='hwd')
dfDicoBuch_new
```

```python
# définir une fonction qui retourne quelque chose
def sizePron(strPron):
    #retourner le nombre de caractère de la prononciation
    
    return len(strPron)
```

```python
# appeler cette fonction sur toutes les lignes de la matrice pour remplir une nouvelle colonne grace à apply
dfDicoBuch_new['sizePron'] = dfDicoBuch_new.apply(lambda x: sizePron(x['pron']),1)

dfDicoBuch_new.sample(n=5)
```
