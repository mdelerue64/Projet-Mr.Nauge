## Parsing

```python
from bs4 import BeautifulSoup
from requests import session
```

```python
def recupMP3(mot):
    urlTarget = "https://www.linguee.fr/anglais-francais/traduction/" + mot + ".html"
    response = requests.get(urlTarget)
    soup = BeautifulSoup(response.text,'html.parser')
    souplettesBody = soup.find('body')
    finded = souplettesBody.findAll('a', {"class" : "audio"})
    listURL = []
    listURLUK = []
    listURLUS = []
    for v in finded:
        listURL.append(v['id'])
    print("1",listURL)
    
    for a in listURL:
        if "EN_US" in a:
            listURLUS.append(a)
        elif "EN_UK" in a:
            listURLUK.append(a)
    print("2",listURLUK)

    debutURL = "https://www.linguee.fr/mp3/"
    idURLUK = debutURL + listURLUK[0]
    idURLUS = debutURL + listURLUS[0]
    URLUK=idURLUK + ".mp3"
    URLUS=idURLUS + ".mp3"
    print(URLUK, URLUS)
```

```python
recupMP3("love")
```
